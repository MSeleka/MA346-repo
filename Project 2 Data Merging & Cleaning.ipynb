```python
import pandas as pd

stats = pd.read_csv(DATA_PATH + '/NBA_Player_stats_2025-26.csv')
salary = pd.read_csv(DATA_PATH + '/NBA_player_salaries_2025-26.csv')

stats
salary
```

```python
# Fix the stats dataset column names
stats.columns = stats.iloc[0]
stats = stats.drop(index=0).reset_index(drop=True)

# Fix the salary dataset column names
salary.columns = salary.iloc[0]
salary = salary.drop(index=0).reset_index(drop=True)

```

```python
# Remove blank rows from salary
salary = salary.dropna(subset=['Player']).reset_index(drop=True)

# Keep only the columns needed for this project
stats = stats[['Player', 'Season', 'PTS', 'AST', 'TRB', 'MP']]
salary = salary[['Player', 'Salary_2025-26']]

# Add the season column to the salary dataset
salary['Season'] = '2025-26'

# Reorder salary columns
salary = salary[['Player', 'Season', 'Salary_2025-26']]

# Standardize merge keys
stats['Player'] = stats['Player'].str.strip()
salary['Player'] = salary['Player'].str.strip()
stats['Season'] = stats['Season'].str.strip()
salary['Season'] = salary['Season'].str.strip()


salary
```

```python
stats
```

Removed duplicates to ensure that each player appears only once per season in the statistics and salary dataset. Some players may have multiple rows due to playing for multiple teams in the same season, which would create duplicate entries during the merge.

```python
stats = stats.drop_duplicates(subset=['Player', 'Season'])
```

```python
salary = salary.drop_duplicates(subset=['Player', 'Season'])
```

## Merge the two data sets

### \(worth 20%\)

The goal of this step is to combine the player statistics dataset with the player salary dataset for the 2025–26 NBA season. This will allow us to analyze the relationship between player performance and salary. An inner join is appropriate because we only want to include players who appear in both datasets, ensuring that each observation contains both performance statistics and salary information.

```python
# Place your code here.  Use more than one cell if needed.

merged_nba = pd.merge(stats, salary, on=['Player', 'Season'], how='inner')

merged_nba
```

The two datasets have now been successfully merged using player name and season as the matching keys. Each row in the merged dataset represents a single player in the 2025–26 NBA season and includes both performance statistics and salary information. Because an inner join was used, only players who appear in both datasets are included in the final result.
